```python
import pandas as pd

df = pd.read_csv('sales_march.csv')  # ファイル名は実際のものに置き換え
print(df.columns)  # 列名一覧を表示
```

```python
# 平均単価を計算
df['Average_price_per_unit'] = df['Net sales'] / df['Net items sold']

# Revenue を計算（今回は Net items sold × 平均単価 = Net sales と同じ）
df['Revenue'] = df['Net items sold'] * df['Average_price_per_unit']

# 確認
df[['Product title','Net items sold','Net sales','Average_price_per_unit','Revenue']].head()
```
